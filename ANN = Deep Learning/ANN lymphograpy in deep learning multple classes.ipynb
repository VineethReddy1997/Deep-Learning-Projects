```python
#features and classes
X = data_lympho.iloc[: , 1:18]
y= data_lympho.iloc[: , 0]
# y.replace({1:0,2:1},inplace = True)
# data_lympho['class'].value_counts()
y.value_counts()
```

```python
sns.countplot(y)
```

```python
#over sampleng the datab in classes

!pip install imblearn
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=10)
x_res,y_res = ros.fit_resample(X,y)
```

```python
sns.countplot(y_res)
```

```python
data_lympho['class'].value_counts()
```

```python
y_res.value_counts()
```

```python
#split the data
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x_res, y_res,  test_size=0.2,random_state=0)
```

```python
Y_train.value_counts()
```

```python
#pre processing the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#import the libraries for model training
from tensorflow.keras.layers import Dense,Dropout
from tensorflow.keras.models import Sequential
```

```python
#training the model
model = Sequential()
model.add(Dense(units =18, activation='relu',kernel_initializer = 'glorot_uniform',input_shape = (17,)))
model.add(Dropout(rate = 0.1))
model.add(Dense(units = 15, activation='relu',kernel_initializer = 'glorot_uniform'))
model.add(Dropout(rate = 0.1))
model.add(Dense(units = 10, activation='relu',kernel_initializer = 'glorot_uniform'))
model.add(Dropout(rate = 0.1))
model.add(Dense(units = 6, activation='relu',kernel_initializer = 'glorot_uniform'))
model.add(Dropout(rate = 0.1))
model.add(Dense(5, activation = 'softmax'))
model.summary()
model.output_shape
model.get_config()
model.get_weights()
```

```python
#model compiling

from tensorflow.keras.losses import SparseCategoricalCrossentropy

model.compile(loss='sparse_categorical_crossentropy', metrics=['accuracy'], optimizer = 'adam')
```

```python
X_train.shape, Y_train.shape, X_test.shape, Y_test.shape
```

```python
#fit  model
model.fit(X_train, Y_train, batch_size=10, epochs=200)
```

```python
y_pred = model.predict(X_test)
```

```python
score = model.evaluate(X_train,Y_train,batch_size=10,)
```
